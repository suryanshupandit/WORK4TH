q27-In Python, functions are "first-class objects," meaning you can pass a function name as an argument to another function.

````python
def math_ops(a, b):
    return a+b, a-b, a*b
````

````python
def check_attendance(roll_no, present_list):
    return "Present" if roll_no in present_list else "Absent"
````

````python
def max_of_three(a, b, c):
    return max(a, b, c)
````

````python
def even_odd(n):
    return "Even" if n % 2 == 0 else "Odd"
````

````python
def count_vc(word):
    v = sum(1 for char in word.lower() if char in 'aeiou')
    return v, len(word) - v
````

````python
def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)
````

````python
def to_upper(word):
    return word.upper()
````

````python
def area_circle(r):
    return 3.14159 * r**2
````

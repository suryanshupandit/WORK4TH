```python
s = "Python"
reversed_s = ""
for char in s:
    reversed_s = char + reversed_s
print(reversed_s)
```

7.Print Fibonacci series up to N terms

```python
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

11.Find the second largest number in a list

```python
lst = [10, 20, 4, 45, 99]
unique_lst = list(set(lst)) # Remove duplicates first
unique_lst.sort()
print(unique_lst[-2])
```

12.Remove duplicates from a list (without set)

```python
lst = [1, 2, 2, 3, 4, 4, 5]
unique = []
for x in lst:
    if x not in unique:
        unique.append(x)
print(unique)
```

13. Count frequency of elements in a list

```python
lst = [1, 2, 1, 3, 2, 1]
freq = {}
for x in lst:
    freq[x] = freq.get(x, 0) + 1
print(freq)
```

19. Flatten a nested list

```python
nested = [[1, 2], [3, 4], [5]]
flat = [item for sublist in nested for item in sublist]
print(flat)
```

20. Find all pairs with a given sum

```python
lst = [1, 2, 3, 4, 5]
target = 5
pairs = []
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] + lst[j] == target:
            pairs.append((lst[i], lst[j]))
print(pairs)
```

21. Print a right triangle pattern

```python
n = 5
for i in range(1, n + 1):
    print("*" * i)
```

22. Print a pyramid pattern

```python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

23. Print all prime numbers in a range

```python
start, end = 10, 30
for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

24. Check if a number is Armstrong

```python
num = 153
order = len(str(num))
sum_val = sum(int(digit) ** order for digit in str(num))
print(num == sum_val)
```
